```
from google.colab.patches import cv2_imshow
import cv2
```

##**Ejemplo a Escala de Grises**

```
img = cv2.imread('Images/88.jpg')
cv2_imshow(img)
```

```
grises = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grises)
```

```
bordes = cv2.Canny(grises, 50, 150)
cv2_imshow(bordes)
```
